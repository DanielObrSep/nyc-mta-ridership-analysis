```python
import pandas as pd
```

```python
df = pd.read_csv("../data_raw/MTA_Daily_Ridership.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("Columnas:", df.columns.tolist())
print(df.columns.tolist)
```

```python
print(df.dtypes)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(':', '')
    .str.replace('-','_')
    .str.replace(' ','_')
    .str.replace('__','_')
)
print(df.columns.tolist)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
print(df.dtypes)
```

```python
print("\nValores nulos por columna:")
print(df.isnull().sum())

print("\nDuplicados exactos:", df.duplicated().sum())
```

```python
df['weekday'] = df['date'].dt.dayofweek
df['is_weekend'] = df['weekday'].isin([5, 6]).astype(int)

df['total_ridership'] = (
    df['subways_total_estimated_ridership'] +
    df['buses_total_estimated_ridership'] +
    df['lirr_total_estimated_ridership'] +
    df['metro_north_total_estimated_ridership'] +
    df['access_a_ride_total_scheduled_trips'] +
    df['staten_island_railway_total_estimated_ridership']
)

df['pct_change_vs_2019'] = (
    df['total_ridership'] /
    df['bridges_and_tunnels_%_of_comparable_pre_pandemic_day']
) * 100
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(df['date'], df['total_ridership'])
plt.title("Total Estimated Ridership Over Time")
plt.xlabel("Date")
plt.ylabel("Total Riders")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mean = df['total_ridership'].mean()
std = df['total_ridership'].std()
df['outlier'] = ((df['total_ridership'] > mean + 3*std) | (df['total_ridership'] < mean - 3*std)).astype(int)

plt.figure(figsize=(12, 5))
plt.plot(df['date'], df['total_ridership'], label='Ridership')
plt.scatter(df[df['outlier'] == 1]['date'], df[df['outlier'] == 1]['total_ridership'], color='red', label='Outliers')
plt.title("Daily Ridership with Outliers")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.to_csv('../nyc_mta_ridership/seeds/mta_daily_ridership.csv', index=False)
print("\n Exportacin lista: mta_daily_ridership.csv")
```
