```python
import duckdb
import pandas as pd
import os
```

```python
db_path = os.path.join('..', 'dev.duckdb')
con = duckdb.connect(db_path)
```

```python
models = [
    "main.stg_mta_daily_ridership",
    "main.stg_mta_daily_ridership_long",
    "main.mrt_daily_summary",
    "main.mrt_weekday_patterns",
    "main.mrt_service_trends",
    "main.mrt_anomaly_flags"
]

for model in models:
    print(f"\n===== {model} =====")
    df = con.execute(f"SELECT * FROM {model} LIMIT 5").df()
    display(df)
    count = con.execute(f"SELECT COUNT(*) AS rows FROM {model}").df()
    print(f"Total rows: {count['rows'][0]}")
```

```python
con.close()
```
